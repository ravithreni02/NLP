```python
corpus="""hello myself ravi's. Welcome to my academy. Please contact me! To become NLP expert."""
```

```python
print(corpus)
```

```python
#Tokenization
#sentence-->paragraph
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
documents=sent_tokenize(corpus)
print(documents)
```

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

```python
#tokenization
#paragragh-->words
#sentence-->words
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```

```python
#classification problem
##comments of product is a positive review or negative review
words=["eating","eaten","eats","writes","writing","programming","programs","finally","finalize"]
```

```python
#porterstemmer
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for word in words:
  print(word+"-->"+stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

```python
#regexpstemmer class
from nltk.stem import RegexpStemmer

# Initialize RegexpStemmer with a pattern
# The pattern 'ing$|e$|able$|s$' removes 'ing', 'e', 'able', or 's' from the end of a word
reg_stemmer = RegexpStemmer('ing$|e$|able$|s$')

# Example usage:
print(reg_stemmer.stem('eating'))
print(reg_stemmer.stem('jumped'))
print(reg_stemmer.stem('adjustable'))
```

```python
reg_stemmer.stem("ingeating")
```

```python
#snowball stemmer
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer("english")
for word in words:
  print(word+"-->"+snowballstemmer.stem(word))
```
